```python
from matplotlib import pyplot as plt
```

```python
times_5 = [15.31,22.18,0.48,15.96,1.11]
precisions_5 = [95.36,95.35,81.19,95.28,95.35]
times_10 = [23.59,41.59,0.49,10.20,1.1]
precisions_10 = [99.88,99.86,86.51,99.74,99.86]
times_76 = [130.27,471.78,0.77,0.42,28.52]
precisions_76 = [99.95,99.86,89.36,99.81,99.92]
```

Para K = 5

```python
fig, ax = plt.subplots()
ax.set(title='Precis達o x Tempo de treinamento para k = 5', xlabel='Tempo de treinamento (s)', ylabel='Precis達o (%)')

ax.plot(times_5[0], precisions_5[0], 'o', label='Random Forest')
ax.plot(times_5[1], precisions_5[1], 'o', label='Grad. Boosting')
ax.plot(times_5[2], precisions_5[2], 'o', label='Gaussian NB')
ax.plot(times_5[3], precisions_5[3], 'o', label='KNN')
ax.plot(times_5[4], precisions_5[4], 'o', label='Dec. Tree')

ax.legend(loc='lower right')
```
